```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
iris
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the Decision Tree Classifier
classifier = DecisionTreeClassifier()
```

```python
# Train the classifier
classifier.fit(X_train, y_train)

```

```python
# Predict on the test set
y_pred = classifier.predict(X_test)
```

```python
# Model evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(classifier, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

**Gini**

```python
'''
The Gini index is a measure of the impurity of a set of data.
It is used in decision trees to determine the best feature to split on at each node.
The Gini index ranges from 0 to 1, where 0 represents a pure dataset (all samples belong to the same class) and
1 represents a completely impure dataset (samples are evenly distributed across all classes).
'''
```

```python
'''
#Samples : The number of samples in a node.

#Class : The target variable that the decision tree is trying to predict.

#Value : The number of samples in each class at a node.
'''
```

**Analysis**

```python
'''
The tree is a decision tree for the Iris dataset, which is a classification dataset with three classes: setosa, versicolor, and virginica.
The tree uses four features to predict the class of a sample: petal length, petal width, sepal length, and sepal width.

The tree starts by splitting on the petal length feature. If the petal length is less than or equal to 2.45 cm, the sample is classified as setosa.
If the petal length is greater than 2.45 cm, the tree splits on the petal width feature.
If the petal width is less than or equal to 1.6 cm,
the sample is classified as versicolor. If the petal width is greater than 1.6 cm, the tree splits on the sepal width feature. If the sepal width is
less than or equal to 3.1 cm, the sample is classified as versicolor. If the sepal width is greater than 3.1 cm, the sample is classified as virginica.

The Gini index at the root node is 0.664, which means that the dataset is fairly impure.
The Gini index decreases at each split, which means that the tree is becoming more pure.
The Gini index at the leaf nodes is 0, which means that all of the samples at each leaf node belong to the same class.

Overall, the tree appears to be a good fit for the Iris dataset. It has a simple structure and it is able to classify the samples accurately.
'''
```

```python
# Gather user input
sepal_length = float(input("Enter sepal length: "))
sepal_width = float(input("Enter sepal width: "))
petal_length = float(input("Enter petal length: "))
petal_width = float(input("Enter petal width: "))

# Make a prediction using the user input
user_input = [[sepal_length, sepal_width, petal_length, petal_width]]
predicted_class = classifier.predict(user_input)

# Map predicted class to the target names
predicted_species = iris.target_names[predicted_class]

# Output the predicted species
print(f"The predicted species is: {predicted_species[0]}")
```
